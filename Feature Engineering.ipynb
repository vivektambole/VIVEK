```python
consistent_team = ['Australia', 'New Zealand', 'England', 'South Africa', 'West Indies', 'Sri Lanka', 'India', 'Bangladesh']
```

```python
df = df[df.batting_team.isin(consistent_team)]
```

```python
df = df[df.bowling_team.isin(consistent_team)]
```

```python
df.sort_values(['start_date', 'innings'], ascending=[True, True])
```

```python
df['overs'] = df['ball']
```

```python
df['ball'] = df['ball'].astype(str)
```

```python
df['ball'].str[0:-1]
```

```python
df['over'] = df['ball'].str[0:-1]
```

```python
df['over'].str.replace('.', '')
```

```python
df['over'] = df['over'].str.replace('.', '')
```

```python
df['ball'] = df['ball'].str[-1]
```

```python
df['total_runs'] = df['runs_off_bat'] + df['extras']
```

```python
df.head()
```

```python
df1 = df[['match_id', 'innings', 'batting_team', 'bowling_team', 'ball', 'overs', 'over', 'total_runs', 'player_dismissed', 'extras']]
```

```python
df1 = df1.replace(np.nan, 0)
```

```python
df1 = df1.rename(columns={'match_id': 'id'})
```

```python
df1.head()
```

```python
df1['total'] = df1.groupby(['id', 'innings'])['total_runs'].transform(sum)
```

```python
df1['total_score'] = df1.groupby(['id', 'innings'])['total_runs'].apply(lambda x: x.cumsum())
```

```python
tmp_1 = df1.groupby(['id', 'innings'])['total_runs'].rolling(min_periods=1, window=30).sum().reset_index()
```

```python
tmp_1[['total_runs']]
```

```python
df1['pre_30_runs'] = tmp_1['total_runs'].to_list()
```

```python
df1['player_dismissed'] = np.where(df1['player_dismissed']==0, 0, 1)
```

```python
tmp_2 = df1.groupby(['id', 'innings'])['player_dismissed'].rolling(min_periods=1, window=30).sum().reset_index()
```

```python
df1['pre_30_wickets'] = tmp_2['player_dismissed'].to_list()
```

```python
df1['wickets_fallen'] = df1.groupby(['id', 'innings'])['player_dismissed'].apply(lambda x: x.cumsum())
```

```python
df1['pre_30_dot_balls'] = df1['total_runs']
```

```python
df1['pre_30_dot_balls'] = np.where(df1['pre_30_dot_balls']==0, 1, 0)
```

```python
tmp_3 = df1.groupby(['id', 'innings'])['pre_30_dot_balls'].rolling(min_periods=1, window=30).sum().reset_index()
```

```python
df1['pre_30_dot_balls'] = tmp_3['pre_30_dot_balls'].to_list()
```

```python
df1['pre_30_boundaries'] = df1['total_runs']
```

```python
df1['pre_30_boundaries'] = np.where(df1['pre_30_boundaries']>3, 1, 0)
```

```python
tmp_4 = df1.groupby(['id', 'innings'])['pre_30_boundaries'].rolling(min_periods=1, window=30).sum().reset_index()
```

```python
df1['pre_30_boundaries'] = tmp_4['pre_30_boundaries'].to_list()
```

```python
tmp_5 = df1.groupby(['id', 'innings'])['extras'].rolling(min_periods=1, window=30).sum().reset_index()
```

```python
df1['pre_30_extras'] = tmp_5['extras'].to_list()
```

```python
df1.head()
```

```python
convert_dict = {'pre_30_boundaries' : int, 'pre_30_dot_balls': int, 'pre_30_wickets': int, 'pre_30_runs': int}

df1 = df1.astype(convert_dict)
```
